```python
import mlflow
from sklearn.datasets import load_iris
```

# Craeating Experiments 

```python
# Create a new experiment

experiment_id=mlflow.create_experiment(
    "example-experiment",
    tags={"topic":"experiment-management", "version":"v1"}
    )
print(f"Experiment created with ID: {experiment_id}")
```

```python
# Set the experiment as active or create it if it doesn't exist 

experiment=mlflow.set_experiment(experiment_name="example-experiment")
print(f"Experiment set to: {experiment.experiment_id}")
```

# Retrieving Experiments

```python
experiment=mlflow.get_experiment_by_name('example_experiment')
if experiment is None:
    print("Experiment not found")
```

```python
experiment=mlflow.get_experiment_by_name("example-experiment")
print(f"Experiment found: {experiment.experiment_id}, {experiment.name}, {experiment.lifecycle_stage}")
```

```python
experiment.to_proto()
```

# Updating Experiments

```python
experiment=mlflow.set_experiment(experiment_name='example-experiment')
print(f"Experiment set to: {experiment.to_proto()}")
```

```python
# create a new tag for the experiment
mlflow.set_experiment_tag(key='new-tag',value='new-tag-value')

# retrieve the experiment by name
experiment=mlflow.set_experiment('example-experiment')
experiment.to_proto()
```

```python
mlflow.set_experiment_tags(tags={'new-tag-2':'new-tag-value-2','new-tag-3':'new-tag-value-3'})

experiment=mlflow.set_experiment(experiment_name='example-experiment')
experiment.to_proto()
```

# Rename Experiment

```python
client=mlflow.MlflowClient()

client.rename_experiment(experiment_id=experiment.experiment_id, new_name='renamed-experiment')

experiment=mlflow.set_experiment(experiment_name='renamed-experiment')
print(experiment.to_proto())
```
